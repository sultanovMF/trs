```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('../results/lab_3/ex_4_1225.txt',sep=' ',header=None)
data = pd.DataFrame(data)

X = data[0]
Y = data[1]
Z = data[2]


fig = plt.figure() 
ax = fig.add_subplot(111, projection='3d') 
ax.plot_trisurf(X, Y, Z, color='white', edgecolors='cornflowerblue', alpha=0.5) 
ax.view_init(elev=30, azim=210, roll=15)
#ax.scatter(X, Y, Z)
plt.show()
```
